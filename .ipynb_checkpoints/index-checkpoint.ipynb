## Data Analysis

**a) How success vary across different studios**

```python
movie_gross_df["domestic_gross"].describe()
```

```python
studio_by_revenue=movie_gross_df.groupby("studio")["domestic_gross"].mean().sort_values(ascending=False)
studio_by_revenue.head(7)
```

```python
#Top studio by domestic gross
studio_by_revenue =movie_gross_df.groupby("studio")["domestic_gross"].mean().sort_values(ascending=False).head(10)

sns.barplot(x=studio_by_revenue.values, y=studio_by_revenue.index, color="brown")
plt.title("Top Studios by Domestic Gross")
plt.xlabel("domestic_gross")
plt.ylabel("studio_names")
plt.show()
```

* We want perform a statistical test to confirm if the domestic revenue differ across the studios.
by comparing their means.
* The appropriete statistical test for this particular analysis is ANOVA TEST, because we are comparing
numeric vs categorical variables between more than 2 categorical groups(studios).

Stating the hypothesis

**Null Hypothesis**: The means of domestic gross across the studios are the same

**Alternative Hypothesis**: The means of in domestic gross across the studios are not the same.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

formula = 'domestic_gross ~ C(studio)'
lm = ols(formula, movie_gross_df).fit()
sm.stats.anova_lm(lm)
```

**Conlusion**
  
Since the p-value is < 0.05 we Reject the Null hypothesis.
    
Hence we agree with the alternative hypothesis that domestic gross means are different across the studios,
meaning there is a strong statistical evidence that studios that produce a movie have an impact on how much money it earns domestically.


**Recommendation**

The company should study BV(Bueno Vista) a division of Disney studios strategy model in production being that it displayed the highest domestic mean gross

**b) Genre trends over-time** 

```python
# spliting strings with 2 or more genres in a string to one
movie_details_df["genres"] = movie_details_df["genres"].apply(lambda x: x.split(",") if isinstance(x,str) else x)
movie_details_df = movie_details_df.explode("genres")

genre_trends = pd.crosstab(movie_details_df["start_year"],movie_details_df["genres"])

genre_trends.plot(figsize=(12, 7))
plt.xlabel("Year")
plt.ylabel("Movies count")
plt.title("Trends in Movie Genres Across The Years")
plt.show()
```

**Observation**

Over the years(2010-2018) Drama production dominated the film industry with over 3,000 movies , followed by Comedy and Documentary 
with over 1500 movies.

c)**Analyzing how the production budget relates with the worldwide gross revenue**

```python
sns.scatterplot(data=movie_budget_df, x="production_budget", y="worldwide_gross")
plt.title("Production Budget vs Worldwide Gross")
plt.xlabel("Production Budget in ($)")
plt.ylabel("Worldwide Gross ($)")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
# target variables
x = movie_budget_df["production_budget"].values.reshape(-1,1)
y = movie_budget_df["worldwide_gross"].values
# spliting the data into test and train sets. 20% test, 80% train
budget_train, budget_test, gross_train, gross_test = train_test_split(x, y, test_size=0.2, random_state=0) 
```

```python
# building the model
from sklearn.linear_model import LinearRegression
# creating the model
model = LinearRegression()
# fiting the model using train data
model.fit(budget_train, gross_train)

#display the intercept (Bias)
print(model.intercept_)

#display the slope (coefficient)
print(model.coef_)
```

```python
# comparing actual and predicted
gross_predict = model.predict(budget_test)
#creating a comparison DataFrame
comparison_frame = pd.DataFrame({"Actual" : gross_test.flatten(), "predicted" : gross_predict.flatten()})
#display stats
comparison_frame.describe()
```

```python
#plotting scatter and regression line
plt.scatter(budget_test, gross_test, color="black")
plt.plot(budget_test, gross_predict, color="red", linewidth=2)
plt.title("Production Budget vs Worldwide Gross Train-Test Linear Regression")
plt.xlabel("Production budget in ($)")
plt.ylabel("Worldwide Gross in ($)")
plt.show()
```

This plot shows simple linear regression between the movies production budget in dollars(independent variable)  and worldwide gross in dollars
(dependent variable), with each black dot representing a movie and  the red line representing the regression line, showing the predicted worldwide gross
in respect to production budget.

**Interpretation**

The slope of the regression line is positive meaning there is a positve relationship between production budget and worldwide gross.
As the production budget increases, the worldwide gross also increases, meaning bigger budget movies result to higher income. 

There is a lot of movie spread at the low-budget (Below 100 million dollars) compared to the high-budget (Above 100 million dollars).
At lower-budgets some films display high gross while others don't. At higher-budgets, gross increases rapidly but still fluctuates in some cases.

From the above plot few spread of dots around the regression line suggest that other factors apart from the production budget i.e genre type are
also affecting the worldwide gross
